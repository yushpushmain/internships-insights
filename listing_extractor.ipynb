```python
import pandas as pd
import re

#read file (simplify internship repo's readme.md)
with open("listings.md", "r", encoding="utf-8") as file:
    lines = file.readlines()


data = [] #dataframe data

for line in lines[74:208]: #list of active roles in the file

    if line.startswith("| **["): # Skip header and separator rows

        parts = line.split("|")
        #extract company. Regex extracts contents of square brackets
        company_match = re.search(r'\[([^\]]+)', parts[1])    
        company = company_match.group(1)
        role = parts[2].strip()
        
        # Extract first application link from the <a href="..."> tag
        match = re.search(r'href="(https://simplify\.jobs[^"]+)"', parts[4])
        link = match.group(1) if match else None
        if (link != None):
            data.append((company, role, link))
    
for line in lines[209:]: #list of inactive roles, the job link is now in the parts[1] column
    if line.startswith("| **["):
        parts = line.split("|")
        company_match = re.search(r'\*\*\[([^\]]+)', parts[1])    
        company = company_match.group(1)
        role = parts[2].strip()
        
        # match = re.search(r'\(https://simplify.jobs.+\)**"', parts[1])
        match = re.search(r'(https://simplify\.jobs[^\)]+)', parts[1])
        link = match.group(0) if match else None
        if (link != None):
            data.append((company, role, link))


# Create DataFrame
df = pd.DataFrame(data, columns=["Company", "Role", "Link"])
df.to_csv("simplify_listings.csv")

pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
df
```

```python
import requests
import json
from bs4 import BeautifulSoup
import time
import csv

# Load your 1200 URLs into this list
urls = [
    "https://simplify.jobs/p/8c19aaed-2fe5-4561-95d0-e1c86fc18eae?utm_source=GHList",
    # ... add the rest here
]

# Output file


# Optional: headers to look more like a browser and avoid being blocked
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')
souptext = soup.get_text()

# Open a CSV to store results
script_tag = soup.find("script", type="application/ld+json")
if script_tag:
    job_data = json.loads(script_tag.string)
    title = soup.find("title").string
    jobmatch = re.search(r"(.+) @", title)
    print(jobmatch.group(1))
    title = job_data.get("title", "N/A")
    raw_description = job_data.get("description", "")
    # Remove HTML tags from description
    description_soup = BeautifulSoup(raw_description, "html.parser")
    description = description_soup.get_text(strip=True)
    
    print(f"Title: {title}")
    print(f"Description:\n{description}")

```

```python
container = soup.find("div", class_= "mt-4")
# print(container.prettify())
# print("\n\n\n")
if container:
    # Find all 'text-left' divs inside that container
    sections = container.find_all("div", class_="text-left", recursive=False)
    for section in sections:
    # Extract section title (from the inner div with class 'text-base font-bold...')
        title_div = section.find("div", class_="mt-3 text-base font-bold text-secondary-400")
        if not title_div:
            continue
        title = title_div.get_text(strip=True)

        print(f"Section: {title}")

        # Extract each <li> in the section
        items = section.find_all("li")
        for li in items:
            print(f"- {li.get_text(strip=True)}")
        print()  # Blank line between sections
```

```python
requirements_heading = soup.find("div", class_="mt-3 text-base font-bold text-secondary-400", string="Requirements")
if requirements_heading:
    # Find the nearest <ul> that contains the bullet points
    ul = requirements_heading.find_next("ul")
    
    if ul:
        bullet_points = [li.get_text(strip=True) for li in ul.find_all("li")]
        print("Requirements:")
        for point in bullet_points:
            print("-", point)
    else:
        print("No bullet list found.")
else:
    print("Requirements heading not found.")
```
