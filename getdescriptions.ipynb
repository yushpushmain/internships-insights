```python
import requests
from bs4 import BeautifulSoup
import csv
import time
import regex as re
def sanitize_filename(name): # Replace invalid characters before naming file
    return re.sub(r'[\\/*?:"<>|]', "_", name)

with open('2024-2025_cs_internship_archive.csv', newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    rows = list(reader)

headers = {"User-Agent": "Mozilla/5.0"}
# for i in range(10):
#     row = rows[i]
for row in rows:
    try:
        res = requests.get(row['Link'], headers=headers)
        soup = BeautifulSoup(res.text, "html.parser")

        filename = sanitize_filename(f"{row['Company']}-{row['Role']}") # filename style: Company-Role
        sections = soup.find_all("div", class_="text-left")
        with open(f"descriptions/{filename}.txt", "w", encoding="utf-8") as file:
            print(row)
            file.write(f"{row['Company']} - {row['Role']}:\n")
            for section in sections: # Simplify generally has three sections: Requirements, Responsibilites, and sometimes, Desired Qualifications
                # Extract section title (from the inner div with class 'text-base font-bold...')
                title_div = section.find("div", class_="mt-3 text-base font-bold text-secondary-400")
                if not title_div:
                    continue
                title = title_div.get_text(strip=True)

                file.write(f"Section: {title}\n")

                # Extract each <li> in the section
                items = section.find_all("li")
                for li in items:
                    file.write(f"- {li.get_text(strip=True)}\n")
                file.write("\n")  # Write a blank line after each section

    except Exception as e:
        print(f"Error scraping {row['Link']}: {e}")

    time.sleep(0.5)
```
