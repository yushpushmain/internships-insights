```python
from sklearn.feature_extraction.text import TfidfVectorizer
import os
from collections import Counter
import string
from nltk.corpus import stopwords
import pandas as pd
```

```python
df = pd.read_csv("keywords.csv")
keywords = set(df["Keyword"].tolist())

documents = []
doc_names = []

for filename in os.listdir("descriptions"):
    if filename.endswith(".txt"):
        with open(os.path.join("descriptions", filename), "r", encoding="utf-8") as f:
            text = f.read().lower()
            documents.append(text)
            doc_names.append(filename)

vectorizer = TfidfVectorizer(vocabulary=keywords, ngram_range=(1, 3), stop_words='english', norm = None)
tfidf_matrix = vectorizer.fit_transform(documents)
#print(tfidf_matrix)
```

```python
df = pd.DataFrame(tfidf_matrix.toarray(), index=doc_names, columns=vectorizer.get_feature_names_out())
df.to_csv("tfidf_matrix_filtered.csv")
df
```

```python
top_n = df.stack().nlargest(10)

for (doc, term), score in top_n.items():
    print(f"{term} in {doc} = TF-IDF: {score:.4f}")
```

```python
import regex as re
documents = []
for filename in os.listdir("descriptions"):
    if filename.endswith(".txt"):
        with open(os.path.join("descriptions", filename), "r", encoding="utf-8") as f:
            documents.append(f.read().lower())
phrase_counts = Counter()

for keyword in keywords:
    # word boundaries using regex
    pattern = r'\b' + re.escape(keyword) + r'\b'
    total_matches = sum(len(re.findall(pattern, doc)) for doc in documents)
    if total_matches > 0:
        phrase_counts[keyword] = total_matches

for phrase, count in phrase_counts.most_common(50):
    print(f"{phrase}: {count}")
```

```python

from sklearn.cluster import KMeans
from sklearn.decomposition import TruncatedSVD
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import Normalizer

lsa = make_pipeline(TruncatedSVD(n_components=150), Normalizer(copy=False))
X_lsa = lsa.fit_transform(tfidf_matrix)
kmeans = KMeans(
    n_clusters=4,
    max_iter=100,
    n_init=1,
)
kmeans.fit(X_lsa)
```

```python
original_space_centroids = lsa[0].inverse_transform(kmeans.cluster_centers_)
order_centroids = original_space_centroids.argsort()[:, ::-1]
terms = vectorizer.get_feature_names_out()
for i in range(4):
    print(f"Cluster {i}: ", end="")
    for ind in order_centroids[i, :10]:
        print(f"{terms[ind]}, ", end="")
    print()
```

```python
from collections import defaultdict

clustered_files = defaultdict(list)
#collect files in each label 
for idx, label in enumerate(kmeans.labels_):
    clustered_files[label].append(doc_names[idx])

#files in each cluster
for i in range(kmeans.n_clusters):
    print(f"\nCluster {i} ({len(clustered_files[i])} files):")
    for name in clustered_files[i]:
        print(f"  - {name}")
```
